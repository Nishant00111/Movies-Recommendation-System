```python
#[{"id": 28, "name": "Action"}, {"id": 12, "name": "Adventure"}, {"id": 14, "name": "Fantasy"}, {"id": 878, "name": "Science Fiction"}]'
# convert this whole info into a list ---
```

```python
# let us make a function to convert each which contains this type of text 

def convert(text):
    L = []
    for i in ast.literal_eval(text):
        L.append(i['name']) 
    return L 
```

```python
movies['genres'] = movies['genres'].apply(convert)
movies.head()
```

```python
movies['keywords'] = movies['keywords'].apply(convert)
movies.head()
```

```python
def convert3(text):
    L = []
    counter = 0
    for i in ast.literal_eval(text):
        if counter < 3:
            L.append(i['name'])
        counter+=1
    return L 
```

```python
movies['cast'] = movies['cast'].apply(convert3)
movies.head()
```

```python
def fetch_director(text): # only extracted the job = director coz we need only director name of movie from crew
    L = []
    for i in ast.literal_eval(text):
        if i['job'] == 'Director':
            L.append(i['name'])
            break
    return L 
```

```python
movies['crew'] = movies['crew'].apply(fetch_director)
```

```python
movies.head()
```

# the column overview is in string so let it be converted to list so we concatenate with other list --

```python
movies['overview'] = movies['overview'].apply(lambda x:x.split())
```

```python
movies.head()
```

# their is one more problem arise that in crew and cast column the name Sam worthington and sam mendes is same so we have to make one text so that remove the space between their names ----

```python
movies['genres'].apply(lambda x:[i.replace(" ","")for i in x])
```

```python
# let it make a function which will do same work as --

def collapse(L):
    L1 = []
    for i in L:
        L1.append(i.replace(" ",""))
    return L1
```

```python
movies['crew'] = movies['crew'].apply(collapse)
movies['genres'] = movies['genres'].apply(collapse)
movies['keywords'] = movies['keywords'].apply(collapse)
```

# MAKE A NEW COLUMN AND MERGE OTHER AND THEN DROP COLUMNS WHICH MERGE INTO NEW COLUMN TAG ---

```python
movies['tags'] = movies['overview'] + movies['genres'] + movies['keywords'] + movies['cast'] + movies['crew']
```

```python
# DROPPING THE COLUMN ---

new_df = movies.drop(columns=['overview','genres','keywords','cast','crew'])
new_df.head()
```

```python
# convert the column tag into a list --

new_df['tags'] = new_df['tags'].apply(lambda x: " ".join(x))
new_df.head(15)
```

# DATA IS PERPARE NOW PERFORM THE CONCEPTS OF DEEP LEARNING LIKE COUNTVECTOR AND OTHER TO TRANSFORM ANF TO READ OUR TECT DATA DATA INTO NUMERIC FORM --

# BUT WAIT WE SEE ,here we have a problm that in our max features of cv the some words have same meaning i:e like actions AND action are same but both present which increse length. AS similar many words is like as same so we remove those words with the help of PorterStemmer function of NLTK -- natural language tool kit as--

```python
import nltk
from nltk.stem.porter import PorterStemmer
ps = PorterStemmer()
```

```python
# for e:g

print(ps.stem('loved'))
print(ps.stem('loving')) # so by stemming the two words will present as one word-
```

```python
def stem(text):
    y = []
    
    for i in text.split():
        y.append(ps.stem(i))
        
    return " ".join(y)
```

```python
new_df['tags'].apply(stem)
```

```python
# COUNT VECTOR IS USED TO MAKE MAX FEATURES (MAX NUMBER OF WORDS )OF 
# THE ALL TAGS IN OUR DATA ----
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(max_features=5000,stop_words='english')
# here we have 5000 features --
```

```python
vectors = cv.fit_transform(new_df['tags']).toarray() # CONVERT INTO ARRAY WITH FITTING AND TRANSFORM
```

```python
# THE UPPER PROBLEM IS SOLVED NY GIVING THE KEY OF SORTING IS INDEX [1] AS 

sorted(list(enumerate(similarity[0])),reverse=True,key=lambda x:x[1])

# THIS WILL SOLVE THE PROBLEM AND WE GET SIMILARITY OF A MOVIE WITH 
# OTHERS IN DECENDING ORDER
```

# now we create a function recommend and choose top 5 similarity of a particular movie --

```python
def recommend(movie):
    index = new_df[new_df['title'] == movie].index[0]
    distances= similarity[index]
    movies_list = sorted(list(enumerate(distances)),reverse=True,key = lambda x: x[1])[1:6]
    for i in movies_list:
        print(new_df.iloc[i[0]].title)
        
    
```

```python
recommend("Avatar")
```
